```python
import os
import zipfile
import pandas as pd

# âœ… ì••ì¶• íŒŒì¼ ê²½ë¡œ
ZIP_FILE = "/home/sophia435256/workspace2/dataset/korea_china_word_bag.zip"
EXTRACT_PATH = "/home/sophia435256/workspace2/dataset/korea_china_word_bag"

# âœ… 1. ZIP ì••ì¶• í•´ì œ
if not os.path.exists(EXTRACT_PATH):  # í´ë”ê°€ ì—†ì„ ë•Œë§Œ ì••ì¶• í•´ì œ
    with zipfile.ZipFile(ZIP_FILE, 'r') as zip_ref:
        zip_ref.extractall(EXTRACT_PATH)
    print(f"âœ… ì••ì¶• í•´ì œ ì™„ë£Œ: {EXTRACT_PATH}")
else:
    print(f"âœ… ì´ë¯¸ ì••ì¶•ì´ í•´ì œë¨: {EXTRACT_PATH}")

# âœ… 2. í´ë” ë° íŒŒì¼ êµ¬ì¡° ì¶œë ¥
print("\nğŸ“‚ ë°ì´í„°ì…‹ ë””ë ‰í† ë¦¬ êµ¬ì¡°:")
for root, dirs, files in os.walk(EXTRACT_PATH):
    level = root.replace(EXTRACT_PATH, "").count(os.sep)
    indent = " " * 4 * level
    print(f"{indent}ğŸ“ {os.path.basename(root)}/")
    for file in files:
        print(f"{indent}    ğŸ“„ {file}")

# âœ… 3. CSV íŒŒì¼ ëª©ë¡ ê°€ì ¸ì˜¤ê¸°
csv_files = []
for root, _, files in os.walk(EXTRACT_PATH):
    for file in files:
        if file.endswith(".csv"):
            csv_files.append(os.path.join(root, file))

print(f"\nğŸ“Œ ì´ {len(csv_files)}ê°œì˜ CSV íŒŒì¼ì´ ë°œê²¬ë¨.")

# âœ… 4. ì²« ë²ˆì§¸ CSV íŒŒì¼ ìƒ˜í”Œ ë°ì´í„° ì¶œë ¥
if csv_files:
    sample_csv = csv_files[0]
    print(f"\nğŸ” ìƒ˜í”Œ ë°ì´í„° íŒŒì¼: {sample_csv}")

    # CSV íŒŒì¼ ë¶ˆëŸ¬ì˜¤ê¸° (UTF-8 ë˜ëŠ” EUC-KR ì¸ì½”ë”© ê°ì§€)
    try:
        df = pd.read_csv(sample_csv, encoding="utf-8-sig")
    except UnicodeDecodeError:
        df = pd.read_csv(sample_csv, encoding="euc-kr")

    # CSV ì»¬ëŸ¼ í™•ì¸
    print("\nğŸ“Œ CSV ì»¬ëŸ¼ ëª©ë¡:")
    print(df.columns.tolist())

    # ì²« 5ê°œ ë°ì´í„° ì¶œë ¥
    print("\nğŸ“Œ ì²« 5ê°œ ìƒ˜í”Œ ë°ì´í„°:")
    print(df.head())
else:
    print("âŒ CSV íŒŒì¼ì´ ì—†ìŠµë‹ˆë‹¤!")
```
